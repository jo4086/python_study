```python
# 파일 생성하기
f = open("C:/doit/새파일.txt", 'w')
f.close() #리소스 닫아주기
```

```python
f = open("C:/doit/새파일.txt", 'w')
for i in range(1,11): # 1~10까지 있는 리스트 생성
    data = f'{i}번째 줄입니다.\n'
    f.write(data)
    f.close
```

```python
# 파일에 새로운 내용 추가
f = open("C:/doit/새파일.txt",'a')
for i in range(11, 20):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# with: 파일을 열고 닫는 것을 자동으로 정리
# f= open("foo.txt", "w")
# f.write('hello')
# f.close()

with open("foo.txt", "w") as f:
    f.write("Life is too short, you need python")
```

```python
f = open("C:/doit/새파일.txt", 'r')
for line in f:
    print(line) # 줄바꿈 문자 \n이 포함
f.close()
```

```python
f = open("C:/doit/새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line) # 줄바꿈 문자 \n이 포함

f.close()
```

```python
f = open("C:/doit/새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    line = line.strip() # 줄 끝의 줄 바꿈 문자(\n)를 제거
    print(line)
f.close()
```
