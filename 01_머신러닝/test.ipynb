```python
import numpy as np

# 기존 데이터
x = np.array([1, 2, 6, 8])  # x 좌표
y = np.array([6, 8, 9, 11])  # y 좌표

# 1. 직선 1 (1 -> 2)와 직선 2 (6 -> 8)의 방정식 계산
# 직선 1: 1 -> 2
slope1 = (y[1] - y[0]) / (x[1] - x[0])  # 기울기
intercept1 = y[0] - slope1 * x[0]       # 절편

# 직선 2: 6 -> 8
slope2 = (y[3] - y[2]) / (x[3] - x[2])  # 기울기
intercept2 = y[2] - slope2 * x[2]       # 절편

# 직선 방정식 확인
print("직선 1 방정식: y = {:.2f}x + {:.2f}".format(slope1, intercept1))
print("직선 2 방정식: y = {:.2f}x + {:.2f}".format(slope2, intercept2))

# 2. 보라색 선(중간점 계산)
# 중간점
x_mid1 = (x[0] + x[1]) / 2  # 1.5
y_mid1 = slope1 * x_mid1 + intercept1  # 직선 1에서의 중간점 y

x_mid2 = (x[2] + x[3]) / 2  # 7
y_mid2 = slope2 * x_mid2 + intercept2  # 직선 2에서의 중간점 y

# 보라색 선의 두 점
purple_line = {
    "start": (x_mid1, y_mid1),  # (1.5, 직선 1 중간점)
    "end": (x_mid2, y_mid2),    # (7, 직선 2 중간점)
}

# 결과 출력
print("보라색 선의 시작점: ({:.2f}, {:.2f})".format(*purple_line["start"]))
print("보라색 선의 끝점: ({:.2f}, {:.2f})".format(*purple_line["end"]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 기존 데이터
x = np.array([1, 2, 6, 8])  # x 좌표
y = np.array([6, 8, 9, 11])  # y 좌표

# 1. 각 구간의 직선 방정식 계산
# 직선 1: 1 -> 2
slope1 = (y[1] - y[0]) / (x[1] - x[0])  # 기울기
intercept1 = y[0] - slope1 * x[0]       # 절편

# 직선 2: 2 -> 6
slope2 = (y[2] - y[1]) / (x[2] - x[1])  # 기울기
intercept2 = y[1] - slope2 * x[1]       # 절편

# 직선 3: 6 -> 8
slope3 = (y[3] - y[2]) / (x[3] - x[2])  # 기울기
intercept3 = y[2] - slope3 * x[2]       # 절편

# 중간점 계산
x_mid1 = (x[0] + x[1]) / 2
y_mid1 = slope1 * x_mid1 + intercept1

x_mid2 = (x[2] + x[3]) / 2
y_mid2 = slope3 * x_mid2 + intercept3

# 시각화
plt.figure(figsize=(8, 6))

# 기존 데이터 포인트
plt.scatter(x, y, color='red', label='Data Points')

# 직선 1: 1 -> 2
x_line1 = np.linspace(x[0], x[1], 100)
y_line1 = slope1 * x_line1 + intercept1
plt.plot(x_line1, y_line1, color='orange', label='Line 1 (1 to 2)')

# 직선 2: 2 -> 6
x_line2 = np.linspace(x[1], x[2], 100)
y_line2 = slope2 * x_line2 + intercept2
plt.plot(x_line2, y_line2, color='blue', label='Line 2 (2 to 6)')

# 직선 3: 6 -> 8
x_line3 = np.linspace(x[2], x[3], 100)
y_line3 = slope3 * x_line3 + intercept3
plt.plot(x_line3, y_line3, color='green', label='Line 3 (6 to 8)')

# 보라색 선 (중간점 연결)
plt.plot([x_mid1, x_mid2], [y_mid1, y_mid2], color='purple', linestyle='--', label='Mid Line')

# 중간점
plt.scatter([x_mid1, x_mid2], [y_mid1, y_mid2], color='blue', label='Mid Points')

# 그래프 표시
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Graph with Line Connections and Mid Line')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 기존 데이터
x = np.array([1, 2, 6, 8])  # x 좌표
y = np.array([6, 8, 9, 11])  # y 좌표

# 각 구간의 직선 방정식 계산
# 직선 1: 1 -> 2
slope1 = (y[1] - y[0]) / (x[1] - x[0])
intercept1 = y[0] - slope1 * x[0]

# 직선 2: 2 -> 6
slope2 = (y[2] - y[1]) / (x[2] - x[1])
intercept2 = y[1] - slope2 * x[1]

# 직선 3: 6 -> 8
slope3 = (y[3] - y[2]) / (x[3] - x[2])
intercept3 = y[2] - slope3 * x[2]

# 곡선 생성 함수
def generate_curve(x_vals, slope, intercept, offset, curvature="convex"):
    """Generate a convex or concave curve based on the slope and intercept."""
    if curvature == "convex":
        return slope * x_vals + intercept + offset * np.exp(-(x_vals - np.mean(x_vals)) ** 2)
    elif curvature == "concave":
        return slope * x_vals + intercept - offset * np.exp(-(x_vals - np.mean(x_vals)) ** 2)

# x 범위 설정
x_range1 = np.linspace(x[0], x[1], 100)
x_range2 = np.linspace(x[1], x[2], 100)
x_range3 = np.linspace(x[2], x[3], 100)

# 외접곡선 (Convex) 계산
convex_curve1 = generate_curve(x_range1, slope1, intercept1, 1, "convex")
convex_curve2 = generate_curve(x_range2, slope2, intercept2, 1, "convex")
convex_curve3 = generate_curve(x_range3, slope3, intercept3, 1, "convex")

# 내접곡선 (Concave) 계산
concave_curve1 = generate_curve(x_range1, slope1, intercept1, 1, "concave")
concave_curve2 = generate_curve(x_range2, slope2, intercept2, 1, "concave")
concave_curve3 = generate_curve(x_range3, slope3, intercept3, 1, "concave")

# 시각화
plt.figure(figsize=(10, 6))

# 기존 데이터 포인트
plt.scatter(x, y, color='red', label='Data Points')

# 외접곡선 (Convex)
plt.plot(x_range1, convex_curve1, color='darkred', label='Convex Curve 1')
plt.plot(x_range2, convex_curve2, color='darkred', label='Convex Curve 2')
plt.plot(x_range3, convex_curve3, color='darkred', label='Convex Curve 3')

# 내접곡선 (Concave)
plt.plot(x_range1, concave_curve1, color='red', linestyle='--', label='Concave Curve 1')
plt.plot(x_range2, concave_curve2, color='red', linestyle='--', label='Concave Curve 2')
plt.plot(x_range3, concave_curve3, color='red', linestyle='--', label='Concave Curve 3')

# 원래 직선 표시
plt.plot(x, y, color='blue', linestyle='-', label='Original Lines')

# 그래프 설정
plt.xlabel('x')
plt.ylabel('y')
plt.title('Convex and Concave Curves for Each Line Segment')
plt.legend()
plt.grid()
plt.show()
```
