```python
class COLUMNS:
    def __init__(self, column_names):
        for idx, name in enumerate(column_names):
            setattr(self, f"col{idx+1}", name)
            setattr(self, name, name)

col = COLUMNS(column_names)

# col0 = col.col0
col1 = col.col1
col2 = col.col2
col3 = col.col3
col4 = col.col4
col5 = col.col5
col6 = col.col6
col7 = col.col7
col8 = col.col8
col9 = col.col9

print(col1)
# print('col0:', col0)
print('col1:', col1)
print('col2:', col2)
print('col3:', col3)
print('col4:', col4)
print('col5:', col5)
print('col6:', col6)
print('col7:', col7)
print('col8:', col8)
print('col9:', col9)
```
