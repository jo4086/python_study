```python
# 3번
class Animal:
    def speak(self):
        return '울음소리'
    
class Dog(Animal):
    def speak(self):
        return '멍멍'

class Cat(Animal):
    def speak(self):
        return '야옹'
    
dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
```

```python
# 4번
class Book:
    def __init__(self, title, author, price, stock ):
        self.title = title
        self.author = author
        self.price = price
        self.stock = stock
    def sell(self, request):
        if request > self.stock:
            return f'재고 부족! 남은 수량: {self.stock}권'
        self.stock -= request
        return self.stock
    def restock(self, request):
        self.stock += request
        return self.stock
    def __str__(self):
        return f'제목: {self.title}, 저자: {self.author}, 가격: {self.price}, 재고: {self.stock}'

book = Book('Python Programming', '김삿갓', 23000, 10)


print(book)
print(f'남은 재고: {book.sell(3)}권')
print(f'추가 재고: {book.restock(5)}권')
print(book)
```

```python
# 5번
class EBook(Book):
    def __init__(self, title, author, price, stock, file_size):
        super().__init__(title, author, price, stock)
        self.file_size = file_size
        self.download_count = 0

    def sell(self, request):
        self.download_count += request
        return self.download_count
    def __str__(self):
        return f'제목: {self.title} | 저자: {self.author} | 가격: {self.price} | 파일 크기: {self.file_size}MB | 다운로드 수: {self.download_count}회'
    
ebook = EBook('learn Python', '김삿갓', 20000, 0, 5.5)
print(ebook)
print(f'다운로드 수: {ebook.sell(1)}회')
print(f'다운로드 수: {ebook.sell(3)}회')
print(ebook)
```
