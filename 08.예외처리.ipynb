## try - except 문

```python
try:
    4 / 0
except ZeroDivisionError as e:
    print(e)
```

## 2. try-finally 문

```python
try:
    a = [1,2]
    print('배열 a:',a[1])
    b = 4/0
    print(b)
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
except IndexError:
    print("인덱싱 할 수 없습니다.")
finally:
    print('무조건 실행됩니다.')
```

```python
try:
    a = [4,0]
    print(a[1])
    b = a[0]/a[1]
    print(b)
except ZeroDivisionError as e:
    print(e)
except IndexError as e:
    print(e)
finally:
    print('무조건 실행됩니다.')
```

```python
try:
    a = [4,1]
    b = a[0]/a[1]
    print(a[1])
    print(b)
except (ZeroDivisionError, IndexError) as e:
    print(e)
finally:
    print('무조건 실행됩니다.')
```
