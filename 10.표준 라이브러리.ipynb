### 5. itertools.zip_longest
- íŒŒì´ì¬ì˜ ë‚´ì¥í•¨ìˆ˜ì˜ `zip()`ê³¼ ë™ì¼í•˜ê²Œ ë¬¶ì–´ì„œ íŠœí”Œë¡œ ë§Œë“¬
- BUT, `zip_longest()` í•¨ìˆ˜ëŠ” `iterables`í•œ ê°ì²´ì˜ ê¸¸ì´ê°€ ì„œë¡œ ë‹¤ë¥´ë©´ ê¸´ ê°ì²´ì˜ ê¸¸ì´ì— ë§ì¶¤

````python
# ìœ ì¹˜ì›ìƒ 5ëª… => ê°„ì‹ ì¤Œ
students = ['ì²«ì§¸', 'ë‘˜ì§¸', 'ì…‹ì§¸', 'ë„·ì§¸', 'ë‹¤ì„¯ì§¸']
snacks = ['ì‚¬íƒ•', 'ì´ˆì½”', 'ì ¤ë¦¬']

result = zip(students, snacks)
print(list(result))

# ê²°ê³¼
# [('ì²«ì§¸', 'ì‚¬íƒ•'), ('ë‘˜ì§¸', 'ì´ˆì½”'), ('ì…‹ì§¸', 'ì ¤ë¦¬')]
# ë„·ì§¸, ë‹¤ì„¯ì§¸ëŠ” ê¸¸ì´ê°€ ë§ì§€ ì•Šì•„ ìŠ¤ë‚µì„ ë°›ì§€ ëª»í•¨ => itertools.zip_longest() ì‚¬ìš©
````

````python
import itertools

students = ['ì²«ì§¸', 'ë‘˜ì§¸', 'ì…‹ì§¸', 'ë„·ì§¸', 'ë‹¤ì„¯ì§¸']
snacks = ['ì‚¬íƒ•', 'ì´ˆì½”', 'ì ¤ë¦¬']

result = itertools.zip_longest(students, snacks)
# ë„·ì§¸, ë‹¤ì„¯ì§¸ì—ëŠ” noneìœ¼ë¡œ ë‚˜ì˜´
print(list(result))

# fillvalue='ê°’' => Noneì— í•´ë‹¹ ê°’ì„ ë„£ìŒ
result = itertools.zip_longest(students, snacks, fillvalue='ìƒˆìš°ê¹¡')
print(list(result))
````

### 8. pickle
pickleì€ ê°ì²´ì˜ í˜•íƒœë¥¼ ê·¸ëŒ€ë¡œ ìœ ì§€í•˜ë©´ì„œ íŒŒì¼ì— ì €ì¥í•˜ê³  ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆê²Œ í•˜ëŠ” ëª¨ë“ˆì´ë‹¤. 

````python
import pickle

# ì €ì¥í•  ë°ì´í„° (Python ê°ì²´)
data = {
    "name": "Alice",
    "age": 30,
    "hobbies": ["reading", "cycling", "hiking"]
}

# 1. ë°ì´í„°ë¥¼ pickle íŒŒì¼ì— ì €ì¥
with open("test.txt", "wb") as file:  # "wb"ëŠ” write binary ëª¨ë“œ
    pickle.dump(data, file)
    print("ë°ì´í„°ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# 2. pickle íŒŒì¼ì—ì„œ ë°ì´í„°ë¥¼ ì½ì–´ì˜¤ê¸°
with open("test.txt", "rb") as file:  # "rb"ëŠ” read binary ëª¨ë“œ
    loaded_data = pickle.load(file)
    print("íŒŒì¼ì—ì„œ ë¡œë“œëœ ë°ì´í„°:", loaded_data)
````

````python
# ë°ì´í„°ë¥¼ ì§ë ¬í™”: í”„ë¡œê·¸ë¨ì—ì„œ ì‚¬ìš©ë˜ëŠ” ê°ì²´(ë°ì´í„° êµ¬ì¡°, ë³€ìˆ˜ ë“±)ë¥¼ ì €ì¥í•˜ê±°ë‚˜ ì „ì†¡í•  ìˆ˜ ìˆëŠ” í˜•ì‹ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •
serialized_data = pickle.dumps(data)
print("ì§ë ¬í™”ëœ ë°ì´í„°:", serialized_data)

# ì§ë ¬í™”ëœ ë°ì´í„°ë¥¼ ì—­ì§ë ¬í™”: ì§ë ¬í™”ëœ ë°ì´í„°ë¥¼ ë‹¤ì‹œ ì›ë˜ì˜ ê°ì²´ë¡œ ë³µì›í•˜ëŠ” ê³¼ì •
deserialized_data = pickle.loads(serialized_data)
print("ì—­ì§ë ¬í™”ëœ ë°ì´í„°:", deserialized_data)
````

````python
import glob

glob.glob(f'{base_path}0*') # ?ëŠ” 1ìë¦¬ ë¬¸ìì—´, *ì€ ì„ì˜ì˜ ê¸¸ì´ì˜ ë¬¸ìì—´ì„ ì˜ë¯¸í•œë‹¤. markë¡œ ì‹œì‘í•˜ëŠ” ëª¨ë“  íŒŒì¼
````

````python
import os
all_path = glob.glob(f'{base_path}*')

folders = [item for item in all_path if os.path.isdir(item)]
files = [item for item in all_path if os.path.isfile(item)]


print('í´ë”:')
for folder in folders:
    print(f"  ğŸ“ '{folder}'")

print('íŒŒì¼:')
for file in files:
    print(f"  ğŸ“„ '{file}'")
````
